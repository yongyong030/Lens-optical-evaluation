```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.stats import norm
from scipy.optimize import curve_fit

dt = pd.read_csv("C:/Users/WTA/Desktop/target/mtf.csv")

dt1 = dt[['lp/mm','before']]
dt2 = dt[['lp/mm','after']]
```

```python
fitting1 = np.poly1d(np.polyfit(dt1['lp/mm'],dt1['before'],3))
line = np.array(dt['lp/mm'])
fitting2 = np.poly1d(np.polyfit(dt2['lp/mm'],dt2['after'],3))
```

```python
print(len(line))
print(len(dt1['before']))
```

```python
plt.rcParams['font.family'] = 'Times New Roman'

fig, ax = plt.subplots(1,1,figsize=(8,5))
ax.scatter(dt1['lp/mm'],dt1['before'],s=5,facecolors='none', edgecolors='darkred',alpha = 0.2)
ax.scatter(dt2['lp/mm'],dt2['after'],s=5,facecolors='none', edgecolors='darkblue',alpha= 0.2,marker='^')

l1, = ax.plot(line, fitting1(line),'darkred',linewidth=2,label = 'Before')
l2, = ax.plot(line, fitting2(line),'darkblue',linewidth=2,label = 'After')
ax.hlines(20, 0, 41, color='orange', linestyle='--', linewidth= 2)

ax.set_xlabel('Spatial Frequency (lp/mm)',fontsize = 14)
ax.set_ylabel('Contrast (%)',fontsize = 14)
ax.set_xlim(min(dt2['lp/mm']),max(dt2['lp/mm']))
ax.set_ylim(0,100)

ax.set_title('MTF',fontsize = 18)
ax.legend([l1,l2],['TCL0.42X-132I-5M','VCM110-36-AL'],edgecolor = 'black')

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True)
plt.show()
```
